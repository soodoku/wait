## Load and Preprocess CA Census Data

```python
# LOAD CA CENSUS DATA: Race, Age, Sex, Commuting, Median Income
ca_race_age_sex_df = pd.read_csv("../data/ca_census_data/ca_age_race_sex_ACSDP5Y2019.DP05_data_with_overlays_2020-12-15T230513.csv")
ca_communiting_characteristics_by_sex_df = pd.read_csv("../data/ca_census_data/ca_commuting_characteristics_by_sex_data_with_overlays_2020-12-24T124644.csv")
ca_median_income_past_12_months_df = pd.read_csv("../data/ca_census_data/ca_median_income_past_12_months_ACSST5Y2019.S1901_data_with_overlays_2020-12-15T192911.csv")
```

```python
# Convert NAME to pure zip code number
ca_race_age_sex_df["ZIP"] = [zip.split()[1] for zip in ca_race_age_sex_df["NAME"]]
ca_communiting_characteristics_by_sex_df["ZIP"] = [zip.split()[1] for zip in ca_communiting_characteristics_by_sex_df["NAME"]]
ca_median_income_past_12_months_df["ZIP"] = [zip.split()[1] for zip in ca_median_income_past_12_months_df["NAME"]]
```

```python
# just keep age, race, sex columns that we want
# ca_race_age_sex

# ZIP

# SEX AND AGE
# DP05_0004E - Sex ratio (males per 100 females)
# DP05_0005PE - percentage estimate of population under 5 yrs old
# DP05_0006PE - percentage estimate of population between 5-9
# DP05_0007PE - percentage estimate of population between 10-14
# DP05_0008PE - percentage estimate of population between 15-19
# DP05_0009PE - percentage estimate of population between 20-24 
# DP05_0010PE - percentage estimate of population between 25-34 
# DP05_0011PE - percentage estimate of population between 35-44 
# DP05_0012PE - percentage estimate of population between 45-54 
# DP05_0013PE - percentage estimate of population between 55-59 
# DP05_0014PE - percentage estimate of population between 60-64 
# DP05_0015PE - percentage estimate of population between 65-74 
# DP05_0016PE - percentage estimate of population between 75-84 
# DP05_0017PE - percentage estimate of population 85+

# RACE
# DP_05_0037PE - percentage of population that is White
# DP_05_0038PE - percentage of population that is Black or African American
# DP_05_0039PE - percentage of population that is American Indian and Alaska Native
# DP_05_0044PE - percentage of population that is Asian
# DP_05_0052PE - percentage of population that is Native Hawaiian and Other Pacific Islander
# DP_05_0071PE - percentage of population that is Hispanic or Latino


ca_race_age_sex_columns = ["ZIP",
                           "DP05_0004E","DP05_0005PE","DP05_0006PE", "DP05_0007PE", "DP05_0008PE",
                           "DP05_0009PE", "DP05_0010PE", "DP05_0011PE", "DP05_0012PE", "DP05_0013PE",
                           "DP05_0014PE", "DP05_0015PE", "DP05_0016PE", "DP05_0017PE",
                           "DP05_0037PE", "DP05_0038PE", "DP05_0039PE", "DP05_0044PE",
                           "DP05_0052PE", "DP05_0071PE"
                          ]



ca_ras_df = ca_race_age_sex_df[ca_race_age_sex_columns]
```

```python
ca_ras_df.head()
```

```python
# just keep census columns that we want
# ca_communiting_characteristics_by_sex_df

# ZIP

# S0801_C01_002E - percent of workers age 16+ that drove a car, truck, or van to work
# S0801_C01_037E - percent of commuters with commute time less than 10 minutes
# S0801_C01_038E - percent of commuters with commute time 10-14 minutes
# S0801_C01_039E - percent of commuters with commute time 15-19 minutes
# S0801_C01_040E - percent of commuters with commute time 20-24 minutes
# S0801_C01_041E - percent of commuters with commute time 25-29 minutes
# S0801_C01_042E - percent of commuters with commute time 30-24 minutes
# S0801_C01_043E - percent of commuters with commute time 35-44 minutes
# S0801_C01_044E - percent of commuters with commute time 45-59 minutes
# S0801_C01_045E - percent of commuters with commute time 60+ minutes
# S0801_C01_046E - mean travel time to work in minutes

ca_communiting_characteristics_by_sex_columns = ["ZIP",
                                                 "S0801_C01_002E", "S0801_C01_037E", "S0801_C01_038E", "S0801_C01_039E", "S0801_C01_040E",
                                                 "S0801_C01_041E", "S0801_C01_042E", "S0801_C01_043E", "S0801_C01_044E", "S0801_C01_045E", "S0801_C01_046E",
                                                ]




ca_commute_df = ca_communiting_characteristics_by_sex_df[ca_communiting_characteristics_by_sex_columns]
```

```python
ca_commute_df.head()
```

```python
# just keep census columns that we want
# ca_median_income_past_12_months_df

# S1901_C01_012E - median household income
# S1901_C01_013 - mean household income

ca_median_income_past_12_months_columns = ["ZIP", "S1901_C01_012E", "S1901_C01_013E"]

ca_income_df = ca_median_income_past_12_months_df[ca_median_income_past_12_months_columns]
```

```python
ca_income_df.head()
```

```python
# rename columns with more descriptive names
# drop first row of each df (header text)

# SEX AND AGE
# DP05_0004E - Sex ratio (males per 100 females)
# DP05_0005PE - percentage estimate of population under 5 yrs old
# DP05_0006PE - percentage estimate of population between 5-9
# DP05_0007PE - percentage estimate of population between 10-14
# DP05_0008PE - percentage estimate of population between 15-19
# DP05_0009PE - percentage estimate of population between 20-24 
# DP05_0010PE - percentage estimate of population between 25-34 
# DP05_0011PE - percentage estimate of population between 35-44 
# DP05_0012PE - percentage estimate of population between 45-54 
# DP05_0013PE - percentage estimate of population between 55-59 
# DP05_0014PE - percentage estimate of population between 60-64 
# DP05_0015PE - percentage estimate of population between 65-74 
# DP05_0016PE - percentage estimate of population between 75-84 
# DP05_0017PE - percentage estimate of population 85+

# RACE
# DP_05_0037PE - percentage of population that is White
# DP_05_0038PE - percentage of population that is Black or African American
# DP_05_0039PE - percentage of population that is American Indian and Alaska Native
# DP_05_0044PE - percentage of population that is Asian
# DP_05_0052PE - percentage of population that is Native Hawaiian and Other Pacific Islander
# DP_05_0071PE - percentage of population that is Hispanic or Latino


ca_ras_rename_columns_dict = {
    "DP05_0004E" : "males_per_100_females",
    "DP05_0005PE" : "p_pop_under_5",
    "DP05_0006PE" : "p_pop_between_5_9",
    "DP05_0007PE" : "p_pop_between_10_14",
    "DP05_0008PE" : "p_pop_between_15_19",
    "DP05_0009PE" : "p_pop_between_20_24",
    "DP05_0010PE" : "p_pop_between_25_34",
    "DP05_0011PE" : "p_pop_between_35_44",
    "DP05_0012PE" : "p_pop_between_45_54",
    "DP05_0013PE" : "p_pop_between_55_59",
    "DP05_0014PE" : "p_pop_between_60_64", 
    "DP05_0015PE" : "p_pop_between_65_74", 
    "DP05_0016PE" : "p_pop_between_75_84", 
    "DP05_0017PE" : "p_pop_over_84",
    "DP05_0037PE" : "p_pop_white", 
    "DP05_0038PE" : "p_pop_black" , 
    "DP05_0039PE" : "p_pop_am_indian_alaskan_native", 
    "DP05_0044PE" : "p_pop_asian",
    "DP05_0052PE" : "p_pop_native_hawaiian_pacific_islander", 
    "DP05_0071PE" : "p_pop_hispanic_latino"
}


ca_ras_df.rename(columns=ca_ras_rename_columns_dict, inplace=True)
ca_ras_df = ca_ras_df.iloc[1:]
ca_ras_df.head()
```

```python
# S0801_C01_002E - percent of workers age 16+ that drove a car, truck, or van to work
# S0801_C01_037E - percent of commuters with commute time less than 10 minutes
# S0801_C01_038E - percent of commuters with commute time 10-14 minutes
# S0801_C01_039E - percent of commuters with commute time 15-19 minutes
# S0801_C01_040E - percent of commuters with commute time 20-24 minutes
# S0801_C01_041E - percent of commuters with commute time 25-29 minutes
# S0801_C01_042E - percent of commuters with commute time 30-34 minutes
# S0801_C01_043E - percent of commuters with commute time 35-44 minutes
# S0801_C01_044E - percent of commuters with commute time 45-59 minutes
# S0801_C01_045E - percent of commuters with commute time 60+ minutes
# S0801_C01_046E - mean travel time to work in minutes

ca_commutting_rename_columns_dict = {
    "S0801_C01_002E" : "p_drive_to_work",
    "S0801_C01_037E" : "p_commute_under_10",
    "S0801_C01_038E" : "p_commute_between_10_14",
    "S0801_C01_039E" : "p_commute_between_15_19",
    "S0801_C01_040E" : "p_commute_between_20_24",
    "S0801_C01_041E" : "p_commute_between_25_29", 
    "S0801_C01_042E" : "p_commute_between_30_34", 
    "S0801_C01_043E" : "p_commute_between_35_44", 
    "S0801_C01_044E" : "p_commute_between_45_59", 
    "S0801_C01_045E" : "p_commute_over_59", 
    "S0801_C01_046E" : "mean_commute_time_minutes"
}

ca_commute_df.rename(columns=ca_commutting_rename_columns_dict, inplace=True)
ca_commute_df = ca_commute_df.iloc[1:]
ca_commute_df.head()
```

```python


# S1901_C01_012E - median household income
# S1901_C01_013 - mean household income

ca_income_rename_columns_dict = {
    "S1901_C01_012E": "median_household_income", 
    "S1901_C01_013E": "mean_household_income" 
}

ca_income_df.rename(columns=ca_income_rename_columns_dict, inplace=True)
ca_income_df = ca_income_df.iloc[1:]
ca_income_df.head()

```

```python
# join all census dfs on ZIP
census_df = pd.merge(ca_ras_df,ca_commute_df,on='ZIP',how='outer')
census_df = pd.merge(census_df, ca_income_df, on="ZIP", how="outer")
census_df.head()
```

```python
# Load DMV data
dmv_df = pd.read_csv("../data/dmv_data_output_12_14_2020.csv", index_col=0)
dmv_descriptive_stats = pd.read_csv("../data/dmv_wait_descriptive_stats.csv", index_col=0)

dmv_df = pd.merge(dmv_df, dmv_descriptive_stats, on="name", how="outer")
dmv_df.rename(columns={" zip": "ZIP"}, inplace=True) # rename to match other df

dmv_df.head()
```

```python
# Left Join DMV data and Census data
census_df["ZIP"] = census_df["ZIP"].astype(str).astype(int) #make sure columns match
combined_df = pd.merge(dmv_df, census_df, on="ZIP", how="left")
```

## Plot Basic Loess

```python
# Plot Loess
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# median household income and average weekday wait
loess_median_income_df = combined_df[["wkday_wait_50", "median_household_income"]]
loess_median_income_df = loess_median_income_df[loess_median_income_df['median_household_income'].notna()] # make sure median household income is non NaN
loess_median_income_df["median_household_income"] = loess_median_income_df["median_household_income"].astype(int)


fig, ax = plt.subplots(figsize=(14, 4))
ax = sns.regplot(x="median_household_income", y="wkday_wait_50", data=loess_median_income_df,
           lowess=True);

ax.set_xlabel("median household income ($)")
ax.set_ylabel("median wait time (minutes)")
ax.set_title("Median Wait Time by Median Household Income")

figure = ax.get_figure()    
figure.savefig('../images/loess_median_household_income_dmv_average_wait.png')
```
